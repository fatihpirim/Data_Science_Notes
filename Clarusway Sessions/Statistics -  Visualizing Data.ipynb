```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
```

```python
x
```

```python
y=x**2
```

```python
plt.plot(x,y,color='red')
```

```python
plt.scatter(x,y);
```
