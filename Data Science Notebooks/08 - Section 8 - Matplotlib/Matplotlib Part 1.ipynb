```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline  
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

# Functional:

```python
plt.plot(x,y)

plt.xlabel('Nums')
plt.ylabel('Nums Squared')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

# Object Oriented:

```python
fig = plt.figure()  # creates object, the object is like a blank canvas

axes = fig.add_axes([0.1,0.1,0.8,0.8])  # sets measurement ratios of graph (the dimensions [left, bottom, width, height])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
test = plt.figure()

axeso = test.add_axes([0,0,1.3,0.6])  

axeso.plot(x,y)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('BIG PLOT')

axes2.plot(x,y)
axes2.set_title('SMALL PLOT')
```
