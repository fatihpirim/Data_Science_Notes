```python
x = np.linspace(0,5,11)
y = x ** 2
```

# Figure Size and DPI (dots per inches)

```python
fig = plt.figure(figsize=(8,2))  # 8 by 2 inches

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2),dpi=100)  # 8 by 2 inches
fig.patch.set_facecolor('white')
axes[0].plot(x,y,'g')
axes[0].set_title('Green')
axes[1].plot(y,x,'r')
axes[1].set_title('Red')
plt.tight_layout()
```

# Adding a Legend

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(5,5),dpi=100) 

axes[0].plot(x,x**2,'g')
axes[0].set_title('Green')


axes[1].plot(x,x**3,'r')
axes[1].set_title('Red')

plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='Bruh Squared')
ax.plot(x,x**3,label='Bruh Cubed')

ax.legend(loc=0)
```

```python
# color, line width, alpha (opacity)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#84c7a2',lw=0.5)
ax.plot(x+.1,y,color='red',lw=0.7)
ax.plot(x+.2,y,color='blue',lw=0.9)
ax.plot(x+.3,y,color='pink',lw=1.9)
ax.plot(x+.7,x**2,color='black',lw=5,alpha=0.5)
```

```python
# line style and Markers
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
fig.patch.set_facecolor('white')
ax.plot(x,y,color='red',lw=3,linestyle='--')
ax.plot(x+0.5,y,color='orange',lw=3,linestyle=':')
ax.plot(x+1,y,color='yellow',lw=3,linestyle='steps')
ax.plot(x+1.5,y,color='green',lw=3,linestyle='-.')
ax.plot(x+2,y,color='blue',lw=3,linestyle='-')
```

```python
fig = plt.figure()
fig.patch.set_facecolor('white')

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='blue',lw=3,linestyle='-',marker='o',markersize=10
       ,markerfacecolor='yellow',markeredgewidth=2,markeredgecolor='gray')
```

```python
# limit
```

```python
fig = plt.figure()
fig.patch.set_facecolor('white')

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='blue',lw=3,linestyle='-',marker='o',markersize=10
       ,markerfacecolor='yellow',markeredgewidth=2,markeredgecolor='gray')

ax.set_xlim([0,3])
ax.set_ylim([0,2])
```
